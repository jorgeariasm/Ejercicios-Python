El equipo directivo de la NHL quiere empezar a almacenar la información del [draft desde 1963 hasta 2022](https://www.kaggle.com/datasets/mattop/nhl-draft-hockey-player-data-1963-2022?resource=download). Ya que actualmente no disponen de la infraestructura para crear una base de datos, quieren empezar almacenando la información en una estructura de carpetas y ficheros, basada en los últimos años y el top N de jugadores del draft por nacionalidad  y año de la siguiente manera:

En el directorio de salida *draft_db*, crearemos una carpeta para cada año *year*, dentro de cada carpeta year, crearemos una carpeta con la nacionalidad del jugador, y dentro de esta una carpeta con el nombre de cada jugador *player*. Finalmente para cada jugador crearemos un fichero info.txt que contendrá el equipo que lo seleccionó en el draft,  su anterior equipo, la edad y la posición en la que juega, con el siguiente formato:

La **función principal** deberá tener 4 parametros de entrada, el *path* del fichero *draft_nhl.csv*, el path de salida, el número de años anteriores que se quieren crear en las carpetas y el número por jugadores por año que se quieren guardar, estos 2 últimos parametros deben ser opcionale y tener como valores por defecto 5 y 5 respectivamente.

````python
import os
import pandas as pd

def crear_carpeta_jugador(path_jugador):
    """
    Función que crea la carpeta del jugador en caso de que no exista
    """
    if not os.path.exists(path_jugador):
        os.mkdir(path_jugador)

def crear_carpeta_nacionalidad(path_nacionalidad):
    """
    Función que crea la carpeta de la nacionalidad en caso de que no exista
    """
    if not os.path.exists(path_nacionalidad):
        os.mkdir(path_nacionalidad)

def crear_carpeta_año(path_año):
    """
    Función que crea la carpeta del año en caso de que no exista
    """
    if not os.path.exists(path_año):
        os.makedirs(path_año)

def crear_archivo_jugador(path_jugador, info_jugador):
    """
    Función que crea el archivo de información del jugador
    """
    with open(path_jugador, 'w') as f:
        f.write(info_jugador)

def funcion_principal_draft(datos_csv, path_salida, años=5, jugadores_año=5):
    """
    Función principal que procesa los datos del draft y los almacena en la estructura de carpetas y archivos
    indicada en el enunciado
    """
    # Filtro los datos de los años que quiero procesar
    año_max = datos_csv['year'].max()
    año_min = año_max - años + 1
    datos_filtrados = datos_csv[datos_csv['year'].between(año_min, año_max)]
    
    # Agrupo los datos por año y nacionalidad
    datos_agrupados = datos_filtrados.groupby(['year', 'nationality'])
    
    # Proceso los grupos
    for grupo, grupos_de_datos in datos_agrupados:
        años, nacionalidad = grupo
        path_año = os.path.join(path_salida, str(años))
        path_nacionalidad = os.path.join(path_año, nacionalidad)
        crear_carpeta_año(path_año)
        crear_carpeta_nacionalidad(path_nacionalidad)
        
        # Ordeno los datos por ranking(overall_pick) y selecciono los N primeros jugadores
        top_jugadores = grupos_de_datos.sort_values('overall_pick').head(jugadores_año).reset_index(drop=True)

        
        # Proceso los jugadores
        for _, jugador in top_jugadores.iterrows():
            nombre_jugador = f"{jugador['player']}"
            path_jugador = os.path.join(path_nacionalidad, nombre_jugador)
            crear_carpeta_jugador(path_jugador)
            
            # Creo la información del jugador
            info_jugador = f"team: {jugador['team']}; Equipo_amateur: {jugador['amateur_team']}; Edad: {jugador['age']}; Posición: {jugador['position']}"
            path_archivo_jugador = os.path.join(path_jugador, 'info.txt')
            crear_archivo_jugador(path_archivo_jugador, info_jugador)

# Paths de entrada y salida
path_entrada = './Data/nhldraft.csv'
path_salida = './draft_db'

# Lectura del archivo de datos
datos_csv = pd.read_csv(path_entrada)

# Procesamiento de los datos
funcion_principal_draft(datos_csv, path_salida, años=5, jugadores_año=5)
````

````python
# Función leer libro

import os
import pandas as pd


def read_book(path_texto):
    """
    Función que lee el archivo.txt y devuelve un dataframe con los capítulos separados.
    """
    with open(path_texto, 'r') as f:
        archivo_txt = f.read()
        
    # Definimos un diccionario con el nombre de los capítulos
    capitulos_dict = {
        'I.': 'A Scandal in Bohemia',
        'II.': 'The Red-Headed League',
        'III.': 'A Case of Identity',
        'IV.': 'The Boscombe Valley Mystery',
        'V.': 'The Five Orange Pips',
        'VI.': 'The Man with the Twisted Lip',
        'VII.': 'The Adventure of the Blue Carbuncle',
        'VIII.': 'The Adventure of the Speckled Band',
        'IX.': 'The Adventure of the Engineer’s Thumb',
        'X.': 'The Adventure of the Noble Bachelor',
        'XI.': 'The Adventure of the Beryl Coronet',
        'XII.': 'The Adventure of the Copper Beeches'
    }
    
    # identificamos los capítulos por número romano
    capitulos = []
    for capitulo_num, capitulo_titulo in capitulos_dict.items():
        capitulo_texto = archivo_txt.split(capitulo_num)[1].split(capitulo_titulo)[0]
        capitulos.append((capitulo_num.strip(), capitulo_titulo.strip(), capitulo_texto.strip()))
    
    capitulos_df = pd.DataFrame(capitulos, columns=['Numero', 'Titulo', 'Texto'])
    return capitulos_df
````

````python
# Comprobación de la función
path_texto = 'Data/The Adventures of Sherlock Holmes.txt'

df_capitulos = read_book(path_texto)
df_capitulos
````

````python
# Función separa capítulos

def split_chapters_into_files(df_capitulos, path_ficheros):
    """
    Función que toma el dataframe y la ruta del fichero donde se guardarán los capítulos.
    Genera en la carpeta indicada los ficheros separados.
    """
    # Crear la carpeta si no existe
    if not os.path.exists(path_ficheros):
        os.makedirs(path_ficheros)

    # Iterar por los capítulos y guardar cada uno en un archivo separado
    for i, row in df_capitulos.iterrows():
        nombre_archivo = "Capítulo{:02d}.txt".format(i+1)
        archivo = os.path.join(path_ficheros, nombre_archivo)
        with open(archivo, 'w') as f:
            f.write(row['Texto'])
````

````python
# Función de compresión

import os
import shutil

def compress_files(path_ficheros, fichero_comp, type='zip'):
    """
    Función de compresión que toma la ruta de los ficheros, una carpeta de salida y
    un formato de compresión (por defecto 'zip'). Si la carpeta de salida no existe,
    la crea. 
    """
    # Verificar si la carpeta de salida existe, y si no, crearla
    if not os.path.exists(os.path.abspath(os.path.dirname(fichero_comp))):
        os.makedirs(os.path.abspath(os.path.dirname(fichero_comp)))
    
    # Comprimir los archivos en el tipo de archivo especificado
    if type == 'zip':
        shutil.make_archive(fichero_comp, 'zip', path_ficheros)
    elif type == 'tar':
        shutil.make_archive(fichero_comp, 'tar', path_ficheros)
    else:
        print('Tipo de archivo no válido.')
````

````python
# Función para conocer el tamaño de archivos

def get_size(path):
    """Obtiene el tamaño en kb de un fichero o carpeta"""
    size = os.path.getsize(path) // 1024  # Tamaño en KB
    if os.path.isdir(path):
        return f"El tamaño de la carpeta '{path}' es {size} KB.", size
    else:
        return f"El tamaño del archivo '{path}' es {size} KB.", size
````
